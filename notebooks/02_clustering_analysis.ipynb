````python
import numpy as np
import os
from pathlib import Path
import sys
sys.path.append('..')

# Chemin vers le fichier de features
features_path = "../data/processed/features_resnet.npy"

# VÃ©rification : le fichier existe-t-il ?
if not os.path.exists(features_path):
    print(f"âŒ Fichier non trouvÃ© : {features_path}")
    print("ğŸ“Œ Deux solutions :")
    print("  1. ExÃ©cutez : make features")
    print("  2. Ou lancez l'extraction depuis la cellule ci-dessous (Re-extraction)")
    raise FileNotFoundError(f"Features non extraites. {features_path} est manquant.")

# Chargement des features
features = np.load(features_path)

print(f"âœ… Features chargÃ©es avec succÃ¨s")
print(f"ğŸ“Š Nombre d'images : {features.shape[0]}")
print(f"ğŸ“Š Dimensions par image : {features.shape[1]} (doit Ãªtre 2048)")
````

````python
# Validation et inspection des features
print("="*60)
print("ğŸ” VALIDATION DES FEATURES")
print("="*60)

# VÃ©rification 1 : Valeurs nulles ou infinies
nan_count = np.isnan(features).sum()
inf_count = np.isinf(features).sum()
print(f"âœ“ Valeurs NaN : {nan_count}")
print(f"âœ“ Valeurs infinies : {inf_count}")

# VÃ©rification 2 : Statistiques globales
mean_val = features.mean()
std_val = features.std()
min_val = features.min()
max_val = features.max()

print(f"âœ“ Moyenne globale : {mean_val:.4f}")
print(f"âœ“ Ã‰cart-type global : {std_val:.4f}")
print(f"âœ“ Min/Max : [{min_val:.4f}, {max_val:.4f}]")

# VÃ©rification 3 : Chaque image a-t-elle un vecteur valide ?
valid_images = np.all(np.isfinite(features), axis=1).sum()
print(f"âœ“ Images avec features valides : {valid_images}/{len(features)}")

if valid_images < len(features):
    print("âš ï¸  ATTENTION : Certaines images ont des features invalides !")
else:
    print("âœ… VALIDATION RÃ‰USSIE : Toutes les features sont valides")
````
